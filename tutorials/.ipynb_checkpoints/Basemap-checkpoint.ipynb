The map is created using the Basemap class, which has many options. Without passing any option, the map has the Plate Carrée projection centered at longitude and latitude = 0

##Managing projections

All maps must have a projection. The projection and its features are all assigned when the object Basemap is created. The way to do it is quite different from other libraries (i.e. GDAL), so understanding this point is very important for working with Basemap.

```python
map = Basemap(projection='aeqd', lon_0 = 10, lat_0 = 50)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

plt.show()
```

The Basemap object has the field __proj4string__ that has the string to be used with __proj4__ for calculating the projection parameters without using Basemap.

```python
print(map.proj4string)
```

##Using epsg to set the projection

Check for epsg codes:

http://spatialreference.org/ref/epsg/gda94-mga-zone-55/

But this one is not supported...

```python
map = Basemap(llcrnrlon=140.75,llcrnrlat=-41.75,urcrnrlon=145.35,urcrnrlat=-37.15, resolution = 'h', epsg=28355)

map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()
plt.show()
```
